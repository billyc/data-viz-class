```python
def convert_CH1903_to_latlong(x,y):
    yNorm = (x - 2600000.0) / 1000000.0
    xNorm = (y - 1200000.0) / 1000000.0;

    longitude10000Sec = \
        2.6779094 + \
        4.728982 * yNorm + \
        0.791484 * yNorm * xNorm + \
        0.1306 * yNorm * pow(xNorm, 2) - \
        0.0436 * pow(yNorm, 3)

    latitude10000Sec = \
        16.9023892 + \
        3.238272 * xNorm - \
        0.270978 * pow(yNorm, 2) - \
        0.002528 * pow(xNorm, 2) - \
        0.0447 * pow(yNorm, 2) * xNorm - \
        0.0140 * pow(xNorm, 3)

    return (longitude10000Sec * 100.0 / 36.0, latitude10000Sec * 100.0 / 36.0)
```
